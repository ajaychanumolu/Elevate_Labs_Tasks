```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load built-in Titanic dataset from seaborn
df = sns.load_dataset('titanic')
print(df.head())
```

```python
print(df.describe())              # Summary for numbers
print(df.describe(include='O'))  # Summary for text data
print(df.isnull().sum())         # Check missing values
```

```python
# Histogram for age
df['age'].plot(kind='hist', bins=20, color='skyblue')
plt.title('Age Distribution')
plt.show()

# Boxplot for fare
sns.boxplot(x=df['fare'])
plt.title('Fare Boxplot')
plt.show()
```

```python
# Only select numeric columns for correlation
numeric_df = df.select_dtypes(include='number')

# Draw the heatmap
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Survival count
sns.countplot(x='survived', data=df)
plt.title('Survival Count')
plt.show()

# Survival by sex
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival by Sex')
plt.show()

# Age distribution by survival
sns.histplot(data=df, x='age', hue='survived', multiple='stack')
plt.title('Age vs Survival')
plt.show()
```
